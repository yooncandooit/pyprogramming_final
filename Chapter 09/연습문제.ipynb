```python
# 문제 1
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    
    if bmi < 18.5:
        category = "저체중"
    elif bmi < 23:
        category = "정상"
    elif bmi < 25:
        category = "과체중"
    else:
        category = "비만"
    
    return bmi, category

bmi, category = calculate_bmi(80, 1.80)
print(f"BMI: {bmi:.2f}")
print(f"범주: {category}")
```

```python
# 문제 2
def calculate_price(price, quantity=1, discount=0, tax_rate=0.1):
    base_amount = price * quantity
    discounted = base_amount * (1 - discount)
    final_price = discounted * (1 + tax_rate)
    return final_price

print(calculate_price(10000))
print(calculate_price(10000, quantity=3))
print(calculate_price(10000, quantity=2, discount=0.2))
print(calculate_price(10000, discount=0.1, tax_rate=0))
```

```python
# 문제 3
def number_stats(*numbers):
    if not numbers:
        print("숫자가 입력되지 않았습니다.")
        return None
    
    return {
        'count': len(numbers),
        'sum': sum(numbers),
        'mean': sum(numbers) / len(numbers),
        'min': min(numbers),
        'max': max(numbers)
    }

print(number_stats(1, 2, 3, 4, 5))
print(number_stats(10, 20))
```

```python
# 문제 4
def create_student_profile(name, student_id, **additional_info):
    profile = "=== 학생 프로필 ===\n"
    profile += f"이름: {name}\n"
    profile += f"학번: {student_id}\n"
    
    korean_labels = {
        'major': '전공',
        'grade': '학년',
        'gpa': '학점',
        'club': '동아리'
    }
    
    for key, value in additional_info.items():
        label = korean_labels.get(key, key)
        profile += f"{label}: {value}\n"
    
    return profile

print(create_student_profile(
    "김철수", 
    "2023001",
    major="컴퓨터공학",
    grade=3,
    gpa=3.8,
    club="코딩동아리"
))
```

```python
# 문제 5
def remove_negatives_inplace(numbers):
    """버전 1: 원본 리스트 수정 - 역순으로 순회하여 인덱스 문제 방지"""
    for i in range(len(numbers) - 1, -1, -1):
        if numbers[i] < 0:
            numbers.pop(i)

def remove_negatives_copy(numbers):
    """버전 2: 새 리스트 반환 - 원본 유지"""
    return [num for num in numbers if num >= 0]

list1 = [1, -2, 3, -4, 5, -6]
remove_negatives_inplace(list1)
print("버전 1 - 원본:", list1)

list2 = [1, -2, 3, -4, 5, -6]
new_list = remove_negatives_copy(list2)
print("버전 2 - 원본:", list2)
print("버전 2 - 새 리스트:", new_list)

# 어떤 상황에서 각 버전을 사용하는 것이 좋을까?
# 버전 1: 메모리 사용을 최소화해야 할 때
# 버전 2: 원본 데이터를 보존해야 할 때
```

```python
# 문제 7
def celsius_to_fahrenheit(c):
    return c * 9/5 + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def celsius_to_kelvin(c):
    return c + 273.15

def kelvin_to_celsius(k):
    return k - 273.15

# 함수를 딕셔너리에 저장
converters = {
    'c_to_f': celsius_to_fahrenheit,
    'f_to_c': fahrenheit_to_celsius,
    'c_to_k': celsius_to_kelvin,
    'k_to_c': kelvin_to_celsius
}

def convert_temperature(value, conversion_type):
    if conversion_type in converters:
        return converters[conversion_type](value)
    else:
        return "지원하지 않는 변환 타입입니다."

print(convert_temperature(100, 'c_to_f'))
print(convert_temperature(32, 'f_to_c'))
print(convert_temperature(0, 'c_to_k'))
```

```python
# 문제 8
products = [
    {'name': 'Laptop', 'price': 1200, 'rating': 4.5, 'stock': 15},
    {'name': 'Mouse', 'price': 25, 'rating': 4.8, 'stock': 50},
    {'name': 'Keyboard', 'price': 75, 'rating': 4.3, 'stock': 30},
    {'name': 'Monitor', 'price': 300, 'rating': 4.7, 'stock': 20},
]

price_sorted = sorted(products, key=lambda x: x['price'])
print("가격 오름차순:", [p['name'] for p in price_sorted])

rating_sorted = sorted(products, key=lambda x: x['rating'], reverse=True)
print("평점 내림차순:", [p['name'] for p in rating_sorted])

high_stock = list(filter(lambda x: x['stock'] >= 30, products))
print("재고 30 이상:", [p['name'] for p in high_stock])

# 딕셔너리 복사와 업데이트
price_increased = list(map(lambda x: {**x, 'price': x['price'] * 1.1}, products))
print("가격 10% 인상:", [(p['name'], p['price']) for p in price_increased])

# 가성비 순 (가격/평점이 낮을수록 좋음)
value_sorted = sorted(products, key=lambda x: x['price'] / x['rating'])
print("가성비 순:", [p['name'] for p in value_sorted])
```

```python
# 문제 9
def fibonacci(n):
    """yield 사용 - 처음 n개의 피보나치 수 생성"""
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

def read_lines(filename):
    """파일을 한 줄씩 읽으며 yield, 빈 줄과 주석은 건너뜀"""
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            for line in f:
                line = line.strip()
                if line and not line.startswith('#'):
                    yield line
    except FileNotFoundError:
        print(f"파일을 찾을 수 없습니다: {filename}")

def infinite_counter(start=0, step=1):
    """무한히 증가하는 카운터 - next()로 호출할 때마다 값 생성"""
    current = start
    while True:
        yield current
        current += step

# 피보나치 테스트
for num in fibonacci(10):
    print(num, end=' ')

print('\n')

# 무한 카운터 테스트 (처음 5개만)
counter = infinite_counter(10, 2)
for _ in range(5):
    print(next(counter))
```

```python
# 문제 10
log_lines = [
    "2024-01-01 10:00:00 INFO User login: user123",
    "2024-01-01 10:05:00 ERROR Database connection failed",
    "2024-01-01 10:10:00 INFO User logout: user123",
    "2024-01-01 10:15:00 WARNING High memory usage: 85%",
    "2024-01-01 10:20:00 ERROR File not found: data.txt",
]

def generate_logs(lines):
    for line in lines:
        yield line

def parse_log(logs):
    for log in logs:  # 로그 문자열을 딕셔너리로 파싱
        parts = log.split(' ', 3)  # 최대 4개 부분으로 분할
        yield {
            'timestamp': f"{parts[0]} {parts[1]}",
            'level': parts[2],
            'message': parts[3]
        }

def filter_by_level(logs, level):
    for log in logs:  # 특정 레벨의 로그만 필터링
        if log['level'] == level:
            yield log

def format_log(logs):
    for log in logs:
        yield f"[{log['timestamp']}] {log['level']}: {log['message']}"

# Generator 체인 - 필요할 때만 데이터 처리하므로 메모리 효율적
logs = generate_logs(log_lines)
parsed = parse_log(logs)
errors = filter_by_level(parsed, 'ERROR')
formatted = format_log(errors)

for log in formatted:
    print(log)
```

```python
# 문제 11
import time

def timer(func):
    """함수 실행 시간을 측정하는 decorator"""
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} 실행 시간: {end_time - start_time:.6f}초")
        return result
    return wrapper

def memoize(func):
    """함수 결과를 캐싱하는 decorator - 동일 인자로 재호출 시 저장된 값을 반환"""
    cache = {}  # 클로저로 유지되는 캐시
    
    def wrapper(*args):
        if args in cache:
            print(f"캐시 반환: {args}")
            return cache[args]
        
        print(f"함수 실행: {args}")
        result = func(*args)
        cache[args] = result  # 저장
        return result
    
    return wrapper

# Decorator 순서!!!: 먼저 memoize로 감싸고 -> 그 다음 timer로 감싸기
@timer
@memoize
def fibonacci_slow(n):
    """재귀적 피보나치 (캐시가 없어서 느린 버전)"""
    if n < 2:
        return n
    return fibonacci_slow(n-1) + fibonacci_slow(n-2)

# 테스트
print(f"결과: {fibonacci_slow(10)}")
print(f"결과: {fibonacci_slow(10)}")  # 캐시에서 즉시 반환
print(f"결과: {fibonacci_slow(15)}")

"""-------------------------------------------------------------------------"""

# 추가 도전: 캐시 크기 제한 추가 - 캐시 히트/미스 출력 - 특정 시간 후 캐시 만료
def advanced_memoize(max_size=100, ttl=None):
    """함수 실행 시간을 측정하는 decorator"""
    def decorator(func):
        cache = {}
        stats = {'hits': 0, 'misses': 0}
        
        def wrapper(*args):
            current_time = time.time()
            
            # TTL 체크: 만료된 캐시 항목 삭제
            if ttl:
                expired_keys = [k for k, v in cache.items() 
                              if current_time - v['time'] > ttl]
                for key in expired_keys:
                    del cache[key]
            
            # 캐시 히트
            if args in cache:
                stats['hits'] += 1
                print(f"캐시 히트! (히트: {stats['hits']}, 미스: {stats['misses']})")
                return cache[args]['value']
            
            # 캐시 미스
            stats['misses'] += 1
            result = func(*args)
            
            # 캐시 크기 제한: FIFO 방식으로 가장 오래된 항목 삭제
            if len(cache) >= max_size:
                oldest_key = next(iter(cache))
                del cache[oldest_key]
            
            # 새 결과 캐싱 (값과 시간 저장)
            cache[args] = {
                'value': result,
                'time': current_time
            }
            
            return result
    
    return decorator

@advanced_memoize(max_size=5, ttl=5)
def expensive_function(n):
    # 무거운 계산 시뮬레이션
    time.sleep(0.1)
    return n ** 2

print(expensive_function(10))
print(expensive_function(10))  # 캐시 히트
```
